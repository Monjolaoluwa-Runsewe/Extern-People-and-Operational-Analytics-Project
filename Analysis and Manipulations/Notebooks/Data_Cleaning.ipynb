```python
import pandas as pd
data = pd.read_excel('/content/Copy of glass_reviews_Monjolaoluwa Runsewe.csv')

data = data.drop(columns=['company_name', 'review_id'])
data = data.drop_duplicates()
data = data.dropna(subset=['summary'])
data = data.dropna(axis=1, how='all')
print(data.head())

data.to_csv('glassdoor_cleaned.csv', index=False)
import pandas as pd

data = pd.read_csv('glassdoor_cleaned.csv')
print("=== SHAPE ===")
print(data.shape)

print("\n=== COLUMN NAMES ===")
print(data.columns.tolist())

# Completeness of key review fields
review_cols = ['review_pros', 'review_cons', 'advice_to_management']

print("\n=== NULL / EMPTY COUNT IN REVIEW FIELDS ===")
for col in review_cols:
    if col in data.columns:
        nulls = data[col].isna().sum()
        empties = (data[col].astype(str).str.strip() == '').sum()
        print(f"  {col}: {nulls} nulls, {empties} empty strings")

# placeholder values
print("\n=== PLACEHOLDER VALUES IN REVIEW FIELDS ===")
placeholders = ['-', 'n/a', 'na', 'none', 'null', 'no', '.', 'N/A']

for col in review_cols:
    if col in data.columns:
        mask = data[col].astype(str).str.strip().str.lower().isin([p.lower() for p in placeholders])
        count = mask.sum()
        print(f"  {col}: {count} placeholder values found")
        if count > 0:
            print(data[mask][[col]].head())

# handling Duplicate rows
print("\n=== DUPLICATE ROWS ===")
dupes = data.duplicated().sum()
print(f"  Total fully duplicate rows: {dupes}")


if all(c in data.columns for c in review_cols):
    dupes_on_text = data.duplicated(subset=review_cols).sum()
    print(f"  Duplicate on review fields only: {dupes_on_text}")

# checking Column name hygiene
print("\n=== COLUMN NAME ISSUES ===")
for col in data.columns:
    issues = []
    if col != col.strip():
        issues.append("leading/trailing whitespace")
    if ' ' in col:
        issues.append("contains spaces")
    if col != col.lower():
        issues.append("not lowercase")
    if issues:
        print(f"  '{col}': {', '.join(issues)}")
    else:
        print(f"  '{col}': OK")


print("\n=== SAMPLE OF REVIEW FIELDS ===")
print(data[review_cols].dropna().sample(min(5, len(data))).to_string())
```
